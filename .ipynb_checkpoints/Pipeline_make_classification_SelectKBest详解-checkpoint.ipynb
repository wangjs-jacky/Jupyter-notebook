# make_classification详解

````python
from sklearn.datasets import make_classification
````

```python 
sklearn.datasets.make_classification(n_samples=100,n_features=20,n_informative=2,n_redundant=2,  n_repeated=0, n_classes=2, n_clusters_per_class=2,weights=None,flip_y=0.01, class_sep=1.0, hypercube=True,shift=0.0, scale=1.0, shuffle=True, random_state=None) 
```
> 功能：生成样本集，通常用于分类算法

- **参数：**

   **n_features :** `特征个数= n_informative（） + n_redundant + n_repeated`

   **n_informative：**多信息特征的个数

   **n_redundant：**冗余信息，informative特征的随机线性组合

   **n_repeated ：**重复信息，随机提取n_informative和n_redundant 特征

   **n_classes：**分类类别

   **n_clusters_per_class ：**某一个类别是由几个cluster构成的

   **weights:** 列表类型，权重比

   **class_sep:**乘以超立方体大小的因子。 较大的值分散了簇/类，并使分类任务更容易。默认为1

   **random_state:** 如果是int，random_state是随机数发生器使用的种子; 如果RandomState实例，random_state是随机数生成器; 如果没有，则随机数生成器是np.random使用的RandomState实例。

````python
X, y = make_classification(n_samples=1000,
                           n_informative=3, 
                           n_redundant=0, 
                           random_state=42,
                           n_classes=3,
                           class_sep=2)
````

````python
print(X.shape)
print(y.shape)
````

- 任意取出两个维度，可以发现，基本上没有区分性

````python
import matplotlib.pyplot as plt 
plt.scatter(X[:,0],X[:,2],c=y)
plt.show()
````

- 通过SelectKBest可以选择出最有表征性的两个维度

# SelectKBest 简单使用


使用` KBest` 需要同时导入SelectKBest函数以及对应的指标

- SelectKBest

- 指标：f_regression ，chi2【卡方】

````python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
````

- `.get_support()`可以返回对应的特征索引

通过下面那张图可以发现确实好多了。【调整class_sep这个属性，可以使得各个类的区分性变得很大，默认值是1,，区分性不是很高】

````python
K_Best = SelectKBest(f_regression,k=2).fit(X,y)
print(K_Best.get_support())
X2 = X[:,K_Best.get_support()]
plt.scatter(X2[:,0],X2[:,1],c=y)
````

# Pipeline的详解

这里选择 SVM 作为第二个管道

- Pipeline的使用点：
  -  导入对应的元组 `("名称",estimator)`
    这里的名称是有用的，因为后面通过`.get_params`可以发现，如svm.SVC其中有个属性C 
    
  - 通过`.set_params`可以后续补充设置，导入时候没有来得及加上的属性，传入的规范为："构建管道时候的名称" + "__" + "对应的属性"
    
    如：`.set_params(svc__C=.1)`

````python
from sklearn import svm
from sklearn.pipeline import Pipeline
````

````python
anova_filter = SelectKBest(f_regression, k=3)
clf = svm.SVC(kernel='linear')
anova_svm = Pipeline([('anova', anova_filter), ('svc', clf)])
anova_svm.set_params(anova__k=10, svc__C=.1).fit(X, y)
prediction = anova_svm.predict(X)
anova_svm.score(X, y)
````

-  提取出管道中的estimator，就不用重新去.fit(X,y)
   - 根据关键字提取：anova_svm['anova']
   
   - 根据只读属性`.named_stops` 
     
    - `anova_svm.named_steps.anova.get_support()`
    
    - `anova_svm.named_steps['anova'].get_support()`
   
   - 根据索引提取
    - `anova_svm[0]` <=> `anova_svm['anova']`
    - `anova_svm[1]` <=> `anova_svm['svc']`
    

````python
# getting the selected features chosen by anova_filter
anova_svm['anova'].get_support()
````

````python
# Another way to get selected features chosen by anova_filter
anova_svm.named_steps.anova.get_support()
````

````python
anova_svm.named_steps['anova'].get_support()
````

````python
anova_svm.named_steps['svc']
````

````python
# Indexing can also be used to extract a sub-pipeline.
sub_pipeline = anova_svm[:1]
print(sub_pipeline)
````

- 当然直接可以对提取出的estimator记性操作

````python
coef = anova_svm[-1].coef_
anova_svm['svc'] is anova_svm[-1]
````

如：原来的大小是（1000,20）经过 SelectKBest 得到（1000,3），但是中途通过`anova_svm.set_params(anova__k=10)`所以是（1000,10），经过SVC后拟合得到的系数大小是（n_class,n_features），因为这里属于多分类问题（3类），所以有三条边界曲线。
