- 蛋糕的直径和价格：

```python
x_train = [[6], [8], [10], [14], [18]]
y_train = [[7], [9], [13], [17.5], [18]]
```

- 导入需要的库函数

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 线性回归
def Linear_predict_y(x_train,y_train,x_plot):
    reg = LinearRegression()
    reg.fit(x_train,y_train)
    y_plot = reg.predict(x_plot)
    return y_plot

def PolynomialFeatures_x(x,degree):
    poly = PolynomialFeatures(degree=degree)
    x_new = poly.fit_transform(x)
    return x_new

degree2 = [1,2,3,5]
color = ['b-','r-','k-','g-'] 

xx = np.linspace(0,25,100)
xx = xx.reshape(xx.shape[0],1)

plt.figure(figsize=(12,6))
for index,degree in enumerate(degree2):
    x_train_new = PolynomialFeatures_x(x_train,degree=degree)
    xx_new = PolynomialFeatures_x(xx,degree=degree)
    yy_new_predict = Linear_predict_y(x_train_new,y_train,xx_new)
    plt.plot(xx,yy_new_predict,color[index])
plt.scatter(x_train,y_train,color = 'r')
plt.axis([0, 25, 0, 25])
plt.xlabel("Diameter")
plt.ylabel("Price")
plt.show()
```
