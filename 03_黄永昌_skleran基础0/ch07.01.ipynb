```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def entropy(px):
    return - (px * np.log2(px))

def gini_impurity(px):
    return px * (1 - px)

def entropy_e(px):
    return - (px * np.log(px))


x = np.linspace(0.01, 1, 100)
plt.figure(figsize=(5, 3), dpi=200)
#plt.title('$Entropy(x) = - P(x) * log_2(P(x))$')
plt.xlim(0, 1.1)
plt.ylim(0, 0.6)
plt.xlabel('P(x)')
plt.ylabel('Entropy')
plt.plot(x, entropy(x), 'r-',label="entropy-log2")

plt.plot(x, entropy_e(x), 'b-',label="entropy-loge");
plt.plot(0.5,entropy_e(0.5),'x')

#plt.ylabel('Gini Impurity')
plt.plot(x, gini_impurity(x), 'k-' , label="Gini Impurity")
plt.plot(0.5,gini_impurity(0.5),'x')
plt.plot(0.5,entropy(0.5),'x')
plt.legend(loc='upper right')
```
